```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X=dataset.iloc[: ,[2, 4]].values
```

```python
print(X)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  Km=KMeans(n_clusters=i, init='k-means++',n_init=4, random_state=21)
  Km.fit(X)
  wcss.append(Km.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel("number of clusters")
plt.ylabel("Wcss valule")
plt.show()

```

```python
km=KMeans(n_clusters=4,init='k-means++',n_init=4,random_state=42)
ymeans=km.fit_predict(X)
```

```python
print(ymeans)
```

```python
plt.scatter(X[ymeans==0, 0],X[ymeans==0, 1],s=100,c='red',label='cluster1')
plt.scatter(X[ymeans==1, 0],X[ymeans==1, 1],s=100,c='magenta',label='cluster2')
plt.scatter(X[ymeans==2, 0],X[ymeans==2, 1],s=100,c='cyan',label='cluster3')
plt.scatter(X[ymeans==3, 0],X[ymeans==3, 1],s=100,c='green',label='cluster4')
plt.scatter(km.cluster_centers_[:, 0],km.cluster_centers_[:, 1],s=100,c='black',label='centroids')
plt.xlabel('age')
plt.ylabel('purchase rating')
plt.legend()
plt.show()
```
